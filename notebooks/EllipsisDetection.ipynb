````python
def sampleUnitCircle(samples):
    #return 2 * np.random.random((samples,2)) - 1
    r = np.random.random(samples)
    theta = 2 * math.pi * np.random.random(samples)
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    return np.array(list(zip(x, y)))
````

````python
# Plot a result of the function
data = sampleUnitCircle(10000)
plt.plot(data[:, 0], data[:, 1], '+')
````
